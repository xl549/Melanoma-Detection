```
import numpy as np
import pandas as pd
import os
import glob
from PIL import Image
```

```
# reading the Meta Data file
metadata = pd.read_csv('HAM10000_metadata.csv')

# resize part 1 
# loop over all
img_dir='/content/gdrive/My Drive/Deep Learning Project/Skin Cancer Detection/dataset/HAM10000_images_part_1'

data_path=os.path.join(img_dir, '*g')
files = glob.glob(data_path)
data=pd.DataFrame()
label = []
count=0

for f1 in files:
  img=Image.open(f1)
  img = img.resize((150,150))
  #print(img.size)
  img_np=np.array(img)
  img_np = img_np.flatten()
  img_mat = np.asmatrix(img_np)
  img_mat = img_mat.astype('float32')/255
  data=data.append(pd.DataFrame(img_mat))
  file_name = os.path.basename(f1) #'ISIC_0028278.jpg'
  image_id = os.path.splitext(file_name)
  label.append(image_id[0])
  count=count+1
  print(count)
  
```

```
# save file - party 1
data_part1=data
data_part1['label']=label
print(data_part1.shape)
```

```
np.save('Image_150_150_scaled_P1.npy',data_part1)
```
